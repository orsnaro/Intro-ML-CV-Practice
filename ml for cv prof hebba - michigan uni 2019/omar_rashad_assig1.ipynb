```python

# implementation of KNN with L1 , L2 distances

import numpy as np

class KNN:
	def __init__(self, k= 3, l : int = 1):
		self.k = k
		self.l = l

	def train(self, x, y):
		self.xtr = x 
		self.ytr = y
		
	def predict(self, x):
		num_test = x.shape[0]
		ypred : np.ndarray = np.zeros(num_test , dtype= self.ytr.dtype)
		
		predict_percent = 0.0
		if self.l == 1 : #using L1 distance
			
			for i in range(num_test):
				# if float(i) / float(num_test) == (predict_percent+10.0) / 100.0 :
				# 	predict_percent += 10
				# 	print ( f"Done Predicting {predict_percent}% ... ")
      
				dist = np.sum(np.absolute(self.xtr - x[i,:]), axis = 1)
				
				dist_tuples = zip( range(len(dist)) , dist ) 
				dist_tuples = sorted(dist_tuples , key= lambda a : a[1] )
				
				freq_arr = np.zeros(10, dtype= int)
			
				for j in range(self.k): 
					label_index = dist_tuples[j][0]
					freq_arr [  self.ytr[ int(label_index) ]  ] += 1 #first element is the org image index
						
				kmin_index = np.argmax(freq_arr)
				ypred[i] = kmin_index
		else : # using L2 distance
			for i in range(num_test):
				# if float(i) / float(num_test) == (predict_percent+10.0) / 100.0 :
				# 	predict_percent += 10
				# 	print ( f"Done Predicting {predict_percent}% ... ")
		
				L2dist = np.sqrt(np.sum( np.power(self.xtr - x[i,:] , 2) , axis= 1))
		
				L2dist_tuples = zip( range(len(L2dist)), L2dist )
				L2dist_tuples = sorted(L2dist_tuples , key= lambda a : a[1] )
				
				freq_arr = np.zeros(10 , dtype=int)
				for j in range(self.k): 
					label_index = L2dist_tuples[j][0]
					freq_arr[ self.ytr[ int(label_index) ] ] += 1 #first element is the org image index
				
    
				kmin_index = np.argmax(freq_arr)
				ypred[i] = kmin_index
		return ypred
```

```python
#using KNN model: Define all KNNs
from tensorflow.keras.datasets import cifar10  # ~160mb

from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score

(xtr, ytr), (xtst, ytst) = cifar10.load_data()

no_tst_imgs = 100
xtr, xval, ytr, yval = train_test_split(xtr, ytr, test_size=no_tst_imgs, random_state=42)




xtr = xtr.reshape(xtr.shape[0],-1) / 255.0
xtr = xtr.reshape(xtr.shape[0],-1) / 255.0
xtst = xtst.reshape(xtst.shape[0],-1) / 255.0
xval = xval.reshape(xval.shape[0],-1) / 255.0


ytst = ytst.squeeze() 
yval = yval.squeeze()
ytr = ytr.squeeze()


k1l1nn_obj = KNN(k= 1, l= 1) 
k1l2nn_obj = KNN(k= 1, l= 2) 

k3l1nn_obj = KNN(k= 3, l= 1) 
k3l2nn_obj = KNN(k= 3, l= 2) 

k5l1nn_obj = KNN(k= 5, l= 1) 
k5l2nn_obj = KNN(k= 5, l= 2) 

```

```python
#k=1 TRAIN
k1l1nn_obj.train(xtr , ytr)
k1l2nn_obj.train(xtr , ytr)

#k=1 PREDICT (eval)
y11pred = k1l1nn_obj.predict(xval)
y12pred = k1l2nn_obj.predict(xval)

print (f"k = 1 L1 ans = {np.sum(y11pred == yval[:])/len(y11pred)}")
print(f"k=1 L1 using accuracy_score= {accuracy_score(y11pred , yval)}")
print (f"k = 1 L2 ans = {np.sum(y12pred == yval[:])/len(y12pred)}")
print(f"k=1 L2 using accuracy_score= {accuracy_score(y12pred , yval)}")
```

```python
#k=3 TRAIN
k3l1nn_obj.train(xtr , ytr)
k3l2nn_obj.train(xtr , ytr)

#k=3 PREDICT (eval)
y31pred = k3l1nn_obj.predict(xval)
y32pred = k3l2nn_obj.predict(xval)

print (f"k = 3 L1 ans = {np.sum(y31pred == yval[:])/len(y31pred)}")
print(f"k=3 L1 using accuracy_score= {accuracy_score(y31pred , yval)}")
print (f"k = 3 L2 ans = {np.sum(y32pred == yval[:])/len(y32pred)}")
print(f"k=3 L2 using accuracy_score= {accuracy_score(y32pred , yval)}")
```

```python
#k=5 TRAIN
k5l1nn_obj.train(xtr , ytr)
k5l2nn_obj.train(xtr , ytr)

#k=1 PREDICT (eval)
y51pred = k5l1nn_obj.predict(xval)
y52pred = k5l2nn_obj.predict(xval)

print (f"k = 5 L1 ans = {np.sum(y51pred == yval[:])/len(y51pred)}")
print(f"k=5 L1 using accuracy_score= {accuracy_score(y51pred , yval)}")
print (f"k = 5 L2 ans = {np.sum(y52pred == yval[:])/len(y52pred)}")
print(f"k=5 L2 using accuracy_score= {accuracy_score(y52pred , yval)}")
```
