#### _Programming:_ **1. write a python code to perform gradient decent to find the value ùë• that gives the minimum value of the function ùëì(ùë•)  =  ùë•    ‚àí  4ùë•  + 3.**

```python
def derivative_f(x):
    return 1 - 8*x

def gradient_descent(initial_x, learning_rate, num_iterations):
    x = initial_x
    
    for _ in range(num_iterations):
        gradient = derivative_f(x)
        x = x - learning_rate * gradient
    
    return x

# Define parameters
init_guess = 0
rate = 0.25
itr_no = 100

# Perform gradient descent
min_x = gradient_descent(init_guess, rate, itr_no)

# Calculate the minimum value of the function
min_value = min_x - 4*min_x**2 + 3

print(f"Best x  approximately = {min_x}")
print(f"minimum value of the function  approximately = {min_value}")
```
